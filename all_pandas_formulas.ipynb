```python
import pandas as pd
import numpy as np

# Original retail dataset
data = {
    'CustomerID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Product': ['Laptop', 'Phone', 'Tablet', 'Tablet', 'Camera', 'Camera', 'Laptop', 'Laptop', 'Phone', 'Phone'],
    'Quantity': [1, 2, 1, 1, 2, 1, 1, 3, 2, 1],
    'Price': [70000, 30000, 25000, 25000, 40000, 18000, 15000, 70000, 32000, 30000],
    'PurchaseDate': ['2023-01-10', '2023-01-12', '2023-01-13', '2023-01-15', '2023-01-16',
                     '2023-01-17', '2023-01-18', '2023-01-19', '2023-01-20', '2023-01-21']
}

df = pd.DataFrame(data)

# Make the entire 'Product' column null
df['Product'] = np.nan

# Make the entire 5th row (index 4) null
df.loc[4] = np.nan

print("Retail DataFrame with one column and one row fully set to NaN:\n")
print(df)
```

```python
# pivot table is used to analyze the data
pd.pivot_table(df,values='Profit',index='Category',aggfunc=np.mean)
```

```python
# Total Sales and Profit by Category and Sub-Category
pivot_table = pd.pivot_table(df,
                              values=['Sales', 'Profit'],
                              index=['Category', 'Sub-Category'],
                              aggfunc='sum')
pivot_table
```

```python
# Total Sales by Category and Sub-Category with Grand Totals
pivot_table = pd.pivot_table(df,
                              values='Sales',
                              index=['Category', 'Sub-Category'],
                              aggfunc='sum',
                              margins=True,
                              margins_name='Grand Total')
pivot_table
```
